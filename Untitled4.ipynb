```python
def load_query(path):
    with open(path) as file:
        return file.read()
```

```python
query = load_query('queries/daily_orders.sql')
daily_orders = pysqldf(query)
```

```python
plt.rcParams.update({'figure.figsize': (17, 3), 'figure.dpi':300})
fig, ax = plt.subplots()
sns.lineplot(data=daily_orders.head(50), x='date', y='num_orders')
plt.grid(linestyle='-', linewidth=0.3)
ax.tick_params(axis='x', rotation=90)
```

```python
query = load_query('queries/base.sql')
df = pysqldf(query)
df.to_csv('csvs/dataset.csv')
```

```python
df.head()
```

```python
features = ['order_count_7_day', 'order_count_30_day']
label = 'label'
test_df = df[df['week'] >= '2016-06-04']
train_df = df[df['week'] < '2016-06-04']
X_train, y_train = train_df[features], train_df[label]
X_test, y_test = test_df[features], test_df[label]
```

```python
reg = XGBRegressor(n_estimators=500, learning_rate=0.01)
reg.fit(X_train, 
        y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        eval_metric='mae')
```

```python
predictions = reg.predict(X_test)
```

```python
test_df = test_df.reset_index().drop('index', axis=1)
test_df['predictions'] = pd.Series(predictions)
```

```python
plt.rcParams.update({'figure.figsize': (17, 3), 'figure.dpi':300})
fig, ax = plt.subplots()
sns.lineplot(data=weekly_orders.tail(50), x='week', y='num_orders')
sns.lineplot(data=test_df, x='week', y='predictions')
plt.grid(linestyle='-', linewidth=0.3)
ax.tick_params(axis='x', rotation=90)
```
